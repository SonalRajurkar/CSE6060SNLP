```python
import spacy
nlp = spacy.load("en_core_web_sm")
example = "Google, a company founded by Larry Page and Sergey Brin in the United States of America "\
+ "has one of the world’s most advanced search engines."
doc=nlp(example)

for ent in doc.ents:
    print(ent.text,ent.label_)
```

```python
for token in doc:
    print(token.text, token.pos_, token.tag_)
```

```python
from spacy import displacy
displacy.render(doc,style='ent')
```

### Importing text from web

```python
#importing from website
import requests 
url = "http://www.gutenberg.org/files/2554/2554-0.txt"
response = requests.get(url) #response is received from the given url
```

```python
type(response)
```

```python
html_page = response.text
```

### Cleaning

```python
#for cleaning the text
from bs4 import BeautifulSoup

soup = BeautifulSoup(html_page, 'html.parser')
type(soup)
```

```python
#cleaned doc 
doc = soup.get_text()
print(doc)
```

### Removing Punctuation and spacing

```python
gutten_nlp = nlp(doc[:99999])   #first 99999 words out of 11lakh words
for token in gutten_nlp:
    if not token.is_punct | token.is_space:  #if punctuation or space is not present print that token
        print(token.text)    
```

### Lemmatization and POS 

```python
""" Lemmitization """ 

for token in gutten_nlp:
    lemma = (token.lemma_)
    print(lemma)
```

```python
""" Part of Speech """ 

for token in gutten_nlp:
    tag_token = (token, token.tag_)
    print(tag_token)
```

### Named Entity Recognition

```python
""" Entity Recognition """

for token in gutten_nlp.ents:
    print(token, token.label_)
```

### Count of verbs and nouns

```python
""" Nouns """
nouns = []
for token in gutten_nlp:   #if fullstop or punctuation is not there and universal pos is noun then adding in nouns list
    if not token.is_stop | token.is_punct and token.pos_ == "NOUN":
        nouns.append(token.text)

        
"""" VERBS """
verbs = []
for token in gutten_nlp:  #if fullstop or punctuation is not there and universal pos is verb then adding in verb list
    if not token.is_stop | token.is_punct and token.pos_ == "VERB":
        verbs.append(token.text)
```

```python
print("Nouns present: ",nouns)
```

```python
print("Number of nouns in text:" ,len(nouns))
```

```python
print("Verbs Present: ",verbs)
```

```python
print("Number of verbs present in text: ",len(verbs))
```

### Finding most common entities

```python
from collections import Counter
kind = []
for entity in gutten_nlp.ents:
    kind.append(entity.label_)
kind_count = Counter(kind)    #Counter stores elements in key value pair as dictionary
top_kind = kind_count.most_common(5)
```
