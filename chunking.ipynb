### Chunking on sentence

```python
 #This tokenizer divides a text into a list of sentences by using an unsupervised algorithm to build a model for abbreviation words, collocations, and words that start sentences. 
from nltk.tokenize import PunktSentenceTokenizer

sentence1="I observed that X had a keen interest and was fully involved in the course when I saw his performances during the Lab sessions, where he would be able to grasp new concepts such as query formation and joins. He has been highly active in the technical scene of our college too with him organizing many events."
custom_sent_tokenizer = PunktSentenceTokenizer(sentence1)

tokenized = custom_sent_tokenizer.tokenize(sentence1)
grammar1="NP:{<RB.?>*<VB.?>*<NNP>+<NN>?}"

#<RB.?>* = "0 or more occurences of any tense of adverb," 

#<VB.?>* = "0 or more occurences of any tense of verb,"

#<NNP>+ = "One or more proper nouns," 

#<NN>? = "zero or one singular noun."
```

```python
for i in tokenized:
    words = nltk.word_tokenize(i)
    tagged = nltk.pos_tag(words)
    chunkParser = nltk.RegexpParser(grammar1)
    chunked = chunkParser.parse(tagged)
    print(chunked)
    chunked.draw()     
```

```python
chunks2=nltk.ne_chunk(chunked)
```

```python
print(chunks2)
```

```python
tagged = nltk.pos_tag(token) #pos tagging of the doc
print(tagged[:200])   #First 200 words
```

```python
grammar1 = "NP:{<DT>?<JJ>+<NN>}"  #NP should be formed whenever it finds optional determiner,followed by any number of adjectives
cp1 = nltk.RegexpParser(grammar1) # and then a noun
```

```python
result = cp1.parse(tagged) #entire doc is parsed through the filter
```

```python
print(result)
```

```python
result.draw()
```

```python
chunks1 = nltk.ne_chunk(tagged) #ne_chunk--> recognize named entities using a classifier, classifier adds category names 
print(chunks1)                 # such as ORGANIZATION, PERSON,GPE
```
